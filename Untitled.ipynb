```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os

image_path = r'input\images.jpg'

if not os.path.exists(image_path):
    print(f"Error: File '{image_path}' not found.")
    exit()

image = Image.open(image_path)
image_array = np.array(image)
gray_image = np.dot(image_array[..., :3], [0.2989, 0.5870, 0.1140])

gray_image_pil = Image.fromarray(gray_image.astype('uint8'), 'L')

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Original Image (RGB)")
plt.imshow(image)
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title("Transformed Image (Grayscale)")
plt.imshow(gray_image_pil, cmap='gray')
plt.axis('off')
plt.tight_layout()
plt.show()

output_path = r'output\gray_image.jpg'
output_folder = os.path.dirname(output_path)
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
gray_image_pil.save(output_path)
print(f"Grayscale image saved to {output_path}")
```
